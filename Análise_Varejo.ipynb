```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel("/content/varejo.xlsx")
```

```python
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique().reset_index()
```

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ", "_")
vendas.groupby("Nome_Departamento").idcompra.nunique().reset_index()
```

```python
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP", "Aplicativo")
```

```python
vendas[vendas["estado"].isnull()]
```

```python
vendas["estado"].fillna("MS", inplace=True)
```

```python
vendas.info()
```

```python
media_preco = vendas["Preço"].mean()
```

```python
vendas["Preço"].fillna(media_preco, inplace = True)
```

```python
vendas.info()
```

```python
vendas[["Preço", "Preço_com_frete"]].describe()
```

```python
preco_errado = vendas.query("Preço > Preço_com_frete")
vendas_certo = vendas.query("Preço < Preço_com_frete")
```

```python
vendas_certo.describe()
```

```python
vendas_certo.head()
```

```python
vendas_certo.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
vendas_certo.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending = False).reset_index()
```

```python
round(vendas_certo.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
```

```python
vendas_certo.groupby("Data").idcompra.nunique().sort_values(ascending = False).reset_index()
```

```python
vendas_certo["mes"] = vendas_certo["Data"].dt.month
```

```python
vendas_certo.head()
```

```python
vendas_certo.groupby("mes").idcompra.nunique().sort_values(ascending = False).reset_index()
```

```python
cliente = pd.read_excel("cliente_varejo.xlsx")
cliente.head()
```

```python
cliente[["idade", "renda"]].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({"renda":"float"})
```

```python
cliente.info()
```

```python
vendas_cliente = vendas_certo.merge(cliente, how = "left", on = "cliente_Log")
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending = False).reset_index(),2)
```

```python
agg_idcanal_renda
```

```python
agg_bandeira_idade = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending = False).reset_index(),2)
```

```python
agg_bandeira_idade.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (7,4))

plt.bar(agg_bandeira_idade["bandeira"],agg_bandeira_idade["idade"], color = "green")

plt.ylabel("Média de Idade")

plt.title("Média de Idade por Bandeira", loc = "center", fontsize = 18, fontweight = "bold", color = "black");
```

```python
agg_idcanal_renda
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idcanal_renda["idcanalvenda"], agg_idcanal_renda["renda"], color = "blue")

plt.ylabel("Média de Renda")

plt.title("Média de Renda por Canal", loc = "center", fontsize = 18, fontweight = "bold", color = "black")
```

```python
vendas_por_data = vendas_certo.groupby("Data").idcompra.nunique().reset_index()
vendas_por_data
```

```python
fig = plt.figure(figsize = (25,2))

plt.xlabel("Data da venda")
plt.ylabel("Quantidade de Vendas")
x = vendas_por_data["Data"]
y = vendas_por_data["idcompra"]

plt.plot(x,y, color = "green", linewidth = "2")
plt.show()
```

```python
agg_dpt_preco = round(vendas_certo.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
```

```python
agg_dpt_preco
```

```python
fig = plt.figure(figsize=(60,20))

plt.bar(agg_dpt_preco["Nome_Departamento"], agg_dpt_preco["Preço_com_frete"], color = "green")

plt.ylabel("Média de Preço")

plt.title("Média de Preço por Departamento", loc = "center", fontsize = 50, fontweight = "bold", color = "black");
```
